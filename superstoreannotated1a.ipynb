### produce dataset showing sum of sales according to each category and each region

```python
dfp2 = df.dropna(subset = ['Profit'])
df125 = dfp2.groupby(['Region', 'Category'])['Sales'].sum()
df126 = df125.unstack(level = 0)
#df121 = df121 + 5000
print(df126)
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax = df126.plot.bar(figsize = (20, 10))
ax.set_ylabel ('Total Sales', size = 25)
ax.set_xlabel ('Category', size = 25)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xticklabels(['Furniture', 'Office Supplies', 'Technology'], Fontsize = 20, rotation = 0)

ax.set_yticklabels(['0', '50,000', '100,000', '150,000', '200,000'], fontsize = 20)
ax.set_yticks([0, 50000, 100000, 150000, 200000])
#ax.set_yticklabels([-10000, -5000, 0, 5000, 10000, 15000], Fontsize = 20)
#ax.set_yticklabels([format(-10000, ','), format(-5000, ','), 0, format(5000, ','), format(10000, ','), format(15000), ','], Fontsize = 20)
#ax.get_xaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.legend(fontsize=20)
plt.title('Total Revenue (2014-2017) for each Category for each Region ', fontsize = 30, weight = 'bold')
plt.show()
```

### produce dataset of sum of sales according to each subcategory and each region

```python
df = pd.read_csv('C:\\Users\\Matthew\\Desktop\\datasets\\office_supplies.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['year'] = df['Order Date'].dt.year
#dfp = df.dropna(subset = ['Profit'])
df120 = df.groupby(['Region', 'Sub-Category'])['Sales'].sum()
df121 = df120.unstack(level = 0)
#df121 = df121 + 5000
print(df121)
```

### graph results showing sum of sales for each region and each subcategory over 4 year period

```python
import matplotlib.pyplot as plt
import numpy as np
ax = df121.plot.bar(figsize = (20, 10))
ax.set_ylabel ('Total Sales', size = 25)
ax.set_xlabel ('Sub-Category', size = 25)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xticklabels(['Accessories', 'Appliances', 'Art', 'Binders', 'Bookcases', 'Chairs',
       'Copiers', 'Envelopes', 'Fasteners', 'Furnishings', 'Labels',
       'Machines', 'Paper', 'Phones', 'Storage', 'Supplies', 'Tables'], Fontsize = 20, rotation = 65)

ax.set_yticklabels(['0', '20,000', '40,000', '60,000', '80,000', '100,000'], fontsize = 20)
ax.set_yticks([0, 20000, 40000, 60000, 80000, 100000])
#ax.set_yticklabels([-10000, -5000, 0, 5000, 10000, 15000], Fontsize = 20)
#ax.set_yticklabels([format(-10000, ','), format(-5000, ','), 0, format(5000, ','), format(10000, ','), format(15000), ','], Fontsize = 20)
#ax.get_xaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.legend(fontsize=20)
plt.title('Total Sales (2014-2017) for each Sub-Category', fontsize = 30, weight = 'bold')
plt.show()
```
